#####------**3**
#####----**7**--4
#####--2--**4**--6
#####8--5--**9**--3

That is, 3 + 7 + 4 + 9 = 23.

Find the maximum total from top to bottom of the triangle below:

NOTE: As there are only 16384 routes, it is possible to solve this problem by trying every route. However, Problem 67, is the same challenge with a triangle containing one-hundred rows; it cannot be solved by brute force, and requires a clever method! ;o)

ماکزیمم حاصل جمع اعداد  از بالا به پایین را در مثلثی که در تصویر می بینید را بیابید:

تذکر:از آنجا که تنها 16384 مسیر برای مثلث فوق از بالا به پایین وجود دارد ، می توان با امتحان کردن هر مسیر این مساله را حل کرد. با این حال ، در تمرین 67 ، همین چالش با مثلثی به شما داده میشود که شامل صد سطر است. در این صورت دیگر نمی توان تمرین را با یک جستجوی فراگیر در تمام مسیر ها  حل کرد ، و نیاز به یک راه حل هوشمندانه تری است ! 😮

````python
triangle = "\
75 \
95 64 \
17 47 82 \
18 35 87 10 \
20 04 82 47 65 \
19 01 23 75 03 34 \
88 02 77 73 07 63 67 \
99 65 04 28 06 16 70 92 \
41 41 26 56 83 40 80 70 33 \
41 48 72 33 47 32 37 16 94 29 \
53 71 44 65 25 43 91 52 97 51 14 \
70 11 33 28 77 73 17 78 39 68 17 57 \
91 71 52 38 17 14 91 43 58 50 27 29 48 \
63 66 04 68 89 53 67 30 73 16 69 87 40 31 \
04 62 98 27 23 09 70 98 73 93 38 53 60 04 23\
"
mynumbers = triangle.split(" ")
mytriangle = {}
i = 0
condition = True
while (i*(i+1)/2)+i < len(mynumbers):
    row = []
    for j in range(0, i+1):
        row += [int(mynumbers[int(i*(i+1)/2)+j])]
    mytriangle[i] = row
    i += 1

paths = ["1", "0"]
for i in range(len(mytriangle)-2):
    mypath = []
    for p in paths:
        mypath.append(p+"0")
        mypath.append(p+"1")
    paths = mypath[:]

paths_and_route = {}

for path in paths:
    i = 0
    j = 0
    route = [mytriangle[j][i]]
    for p in path:
        j += 1
        i += int(p)
        route.append(mytriangle[j][i])
    paths_and_route[path] = (route, sum(route))
maxsum = 0

for value in paths_and_route.values():
    mysum = value[1]
    if mysum > maxsum:
        maxsum = mysum
print(maxsum)
````
