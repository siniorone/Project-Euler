### Highly divisible triangular number
### Problem 12
The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

##### 1: 1
##### 3: 1,3
##### 6: 1,2,3,6
##### 10: 1,2,5,10
##### 15: 1,3,5,15
##### 21: 1,3,7,21
##### 28: 1,2,4,7,14,28
We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

##### به عنوان مثال هفتمین عدد مثلثی  1 + 2 + 3 + 4 + 5 + 6 + 7 = 28 خواهد بود. 
ده عبارت اول عبارتند از:

1 ، 3 ، 6 ، 10 ، 15 ، 21 ، 28 ، 36 ، 45 ، 55 ، ...

اجازه دهید عوامل هفت جمله اول دنباله اعداد مثلثی را لیست کنیم:

##### 1: 1
##### 3: 1,3
##### 6: 1,2,3,6
##### 10: 1,2,5,10
##### 15: 1,3,5,15
##### 21: 1,3,7,21
##### 28: 1,2,4,7,14,28
می بینیم که 28 اولین عدد مثلثی است که بیش از پنج مقسوم علیه دارد.

 اولین  عدد مثلثی که دارای بیش از پانصد مقسوم علیه است را بیابید؟


```python
from sympy.ntheory import factorint

def triangle_number(n):
    return int(n*(n+1)/2)
# Main--------------------------------------------------
triangles = map(triangle_number, range(2, int(10e100)))
condition = True
while condition:
    mytriangle = next(triangles)
    mytriangle_prime_factors = factorint(mytriangle)
    divisors_counter = 1
    for value in mytriangle_prime_factors.values():
        divisors_counter *= (value+1)
    if divisors_counter > 500:
        print(mytriangle)
        break
```
