```python
#------------------------------------------------
def find(s, ch):
    return [i for i, ltr in enumerate(s) if ltr == ch]
#------------------------------------------------
def mypermutations(n):
    str_n=str(n)
    len_n=len(str_n)
    result=[]
    for i in range(len_n):
        result.append((i,))
    for i in range(0,10):
        f=find(str_n,str(i))
        if f !=[] :
            result.append(tuple(f))
    return result  
    
#------------------------------------------------
def change_by_permutation(n,permutations):
     result=[]
     for pp in permutations:
         myr=set()
         for i in range(0,10):
             n_list=list(str(n))
             condition=True
             for p in pp:
                 if p==0 and i==0:
                      condition=False
                 n_list[p]=str(i)
             if condition:myr.add(int("".join(n_list)))
         if myr not in result:
             result.append(myr)           
     return result
#------------------------------------------------
def find_prime_family(n,primes):
      max_count=0
      max_digit=0
      str_n=str(n)
      len_n=len(str_n)

      p=mypermutations(n)
      all_list=change_by_permutation(n,p)
      for all in all_list:
          count=0
          for a in all:
              if a in primes:
                  count += 1
          if count>max_count: 
              max_count=count
      return max_count

# Find All primes number below 1e6---------------
n = int(1e6)
nonprimes=set()
primes = set()
primes_list=[]
for i in range(2,n):
    if i in nonprimes:continue
    else:
        primes.add(i)
        for x in range(i+i, n,i):
            if x not in nonprimes: nonprimes.add(x)
primes_list=list(primes)
primes_list.sort()
# ---------------------------------------------
for prime in primes_list:
    f=find_prime_family(prime,primes)
    if f== 8: 
         print(prime ,f)
         break
```
