```python
def check(*arg):
    if sum(arg) == 200 :
        return True
    return False
n=200
c=0
for p200 in range(0,n+1,200):
    if check(p200):c+=1;break
    for p100 in range(0,n+1,100):
        if check(p200,p100):c+=1;break
        for p50 in range(0,n+1,50):
            if check(p200,p100,p50):c+=1;break
            for p20 in range(0,n+1,20):
                if check(p100,p200,p50):c+=1;break
                for p10 in range(0,n+1,10):
                    if check(p200,p100,p50,p20,p10):c+=1;break
                    for p5 in range(0,n+1,5):
                        if check(p200,p100,p50,p20,p10,p5):c+=1;break
                        for p2 in range(0,n+1,2):
                            if check(p200,p100,p50,p20,p10,p5,p2):c+=1;break
                            p1=200-(p200+p100+p50+p20+p10+p5+p2)
                            if p1>0:c+=1                       
print(c)
```
