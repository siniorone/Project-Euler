```python
def reverse(n):
    str_n = str(n)
    my_n = int(str_n[-1::-1])
    return my_n
def palindrome(n):
    str_n = str(n)
    len_n = len(str_n)
    if len_n == 2 and str_n[0] == str_n[1] :
        return True
    elif len_n > 2 :
        for j in range(0,int(len_n/2)):
            if str_n[j] != str_n[-j - 1] :
                 return False
        return True
    return False

def lychrel(n):
    for z in range(50):
        r = reverse(n)
        if palindrome(n + r) : 
            return False
        else : 
            n = n + r
    return True
count = 0
for i in range(1, 10000):
        lx=lychrel(i)
        if lx: 
            count += 1
print(count)
```
