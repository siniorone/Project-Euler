```python
consecutive=[]
n = int(1e6)
nonprimes=set()
primes = []
i=2
while True:
    # Find All primes number below 1e6---------------
    is_prime=False
    if i in nonprimes:pass
    else:
        is_prime=True
        primes.append(i)
        for x in range(i+i, n,i):
            if x not in nonprimes: nonprimes.add(x)
    # ---------------------------------------------
    if not is_prime:
        x=i
        distinct_prime=0
        for p in primes:
            if p > x : break
            first=True
            while x % p == 0:
                if first :
                    first = False
                    distinct_prime += 1
                x=int(x / p)
    else:distinct_prime=0
    # ---------------------------------------------
    if distinct_prime == 4  :
        consecutive.append((i,distinct_prime))
        if len(consecutive)>=4: 
            print(consecutive)
            print(consecutive[0][0])
            break
    else:
        consecutive=[]
    # ---------------------------------------------   
    i+=1 
```
