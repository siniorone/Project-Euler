```python
def is_prime(n):
    p = 2
    while p < n**.5 :
        if n % p: p +=1 
        else : return False
    return True
primes = set()            
square = set()
i=1
while True:
    i+=1
    square.add(2*(i**2))
    # Prime calculator -----------------------------
    if is_prime(i): primes.add(i)
    # Main calculator ------------------------------            
    if i % 2 != 0  and i not in primes:
        conjecture=False
        for p in primes:
            d=i-p
            if d in square:
                conjecture=True
                break
        if not conjecture:
            print(f"Conjecture is False for {i} ")
            break
        
```
