 <div dir='rtl' align='right'>
حاصلضرب ضرایب معادله فوق  a و b را بیابید به نحوی که معادله درجه دوم فوق بیشترین تعداد اعداد اول را برای اعداد صحیح متوالی با شروع از صفر  (n=0)   تولید کند.
</div>

```python
all_possible_combination = []
quadratic_dict = {"a":0,"b":0 ,"n":0,"len":0}
def is_prime(n):
    for i in range (2, int(n**0.5) + 1):
        if n % i == 0 : return False
    return True
for a in range(-999, 1000, 1):
    for b in range(-1000, 1001, 1):
        n = 0
        detected_prime = set()
        while True:
            if len(detected_prime) == quadratic_dict["len"]:
                if quadratic_dict not in all_possible_combination :
                    all_possible_combination.append(quadratic_dict)
            if len(detected_prime) > quadratic_dict["len"]:
                quadratic_dict["n"] = n
                quadratic_dict["a"] = a
                quadratic_dict["b"] = b
                quadratic_dict["len"] = len(detected_prime)
                all_possible_combination = []
                all_possible_combination.append(quadratic_dict)
            x = n**2 + a*n + b
            if x > 1 and is_prime(x):
                detected_prime.add(x)
            else:
                break
            n += 1
print(all_possible_combination)
print(all_possible_combination[0]['a'] * all_possible_combination[0]['b'])
```
