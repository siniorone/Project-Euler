```python
# Find All primes number below 1e6---------------
n = int(1e6)
nonprimes=set()
primes = set()
for i in range(2,n):
    if i in nonprimes:continue
    else:
        primes.add(i)
        for x in range(i+i, n,i):
            if x not in nonprimes: nonprimes.add(x)
# ------------------------------------------------
consecutive_prime=[]
prime_list=list(primes)
prime_list.sort()
count=0
max_sum=0
for i in range(21,len(prime_list)):
    x=sum(prime_list[0:i])
    if x > n :break
    for j in range(0,len(prime_list)-i+1):
        if j>0:
            x-=prime_list[j-1]
            x+=prime_list[j+i-1]

        if x>n :break
        if x in primes:
            max_sum=x
            count=i
    
print(max_sum,count)
```
