```python
# Function to calculate alphabetical value for each name:
def alphabetical_value(string):
    value = 0
    string = string.upper()
    for s in string:
        value += ord(s)
    return (value - 64*len(string))
# Reading the txt file and store it in a list
f = open("dataset/p042_words.txt", "r")
words = f.read()
words_list = words.split('","')
words_list[0] = words_list[0][1:]  # some modification
words_list[-1] = words_list[-1][:-1]  # some modification
limit=20*90
triangles=set()
tn=0
n=1
while tn <= limit:
    tn = n * (n+1) / 2
    triangles.add(tn)
    n += 1
count = 0
for word in words_list:
    if alphabetical_value(word) in triangles:
        count +=1    
print(count)
```
