```python
import time
start=time.time()
max=0
for i in range(99,0,-1):
    for j in range(99,0,-1):  
        number=i**j
        str_number = str(number)
        mydigits=[int(x) for x in str_number ]
        mysum=sum(mydigits)
        if mysum > max : max = mysum
        if int(max/9) > len(str_number) : break
print(max)
print(time.time()-start)
```
