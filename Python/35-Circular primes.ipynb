```python
def circular(n):
    str_n=str(n)
    l=len(str_n)
    result=set()
    for i in range(l):
        r=str_n[i:]+str_n[0:i]
        result.add(int(r))
    return result
# Find All primes number below 1e6---------------
n = int(1e6)
nonprimes=set()
primes = set()
for i in range(2,n):
    if i in nonprimes:continue
    else:
        primes.add(i)
        for x in range(i+i, n,i):
            if x not in nonprimes: nonprimes.add(x)
# ---------------------------------------------
circular_primes=set()
for prime in primes:
    is_circular_prime=True
    for c in circular(prime):
        if c not in primes:
            is_circular_prime=False
            break
    if is_circular_prime : circular_primes.add(prime)
print(len(circular_primes))
```
