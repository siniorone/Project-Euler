```python
def is_pandigital(a,b,p):
    if len(str(a)+str(b)+str(p))!=9: return False
    digits=set(range(1,10))
    for aa in str(a):
        if int(aa) in digits : digits.remove(int(aa))
    for bb in str(b):
        if int(bb) in digits : digits.remove(int(bb))
    for pp in str(p):
        if int(pp) in digits : digits.remove(int(pp))
    if digits==set() : return True
    return False
pandigitals=set()
for a in range(1,10):
    for b in range(1000,10000):
        p = a * b
        if p not in pandigitals and is_pandigital(a ,b ,p):
            print(f"{a}×{b}={p} is pandigital")
            pandigitals.add(p)
for a in range(10,100):
    for b in range(100,1000):
        p = a * b
        if p not in pandigitals and is_pandigital(a ,b ,p):
            print(f"{a}×{b}={p} is pandigital")
            pandigitals.add(p)
print(pandigitals)
print(sum(pandigitals))
            
```
