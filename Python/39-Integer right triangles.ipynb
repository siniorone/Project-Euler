```python
radical={}
power={}
for i in range(1,1001):
    x=i**2
    radical[x]=i
    power[i]=x
triangles={}
for a in range(1,1000):
    for b in range(1,1000):
        c2=a**2 + b**2
        if c2 in radical:
            c=radical[c2]
            p=a+b+c
            if p <= 1000:
                t_list=[a,b,c]
                t_list.sort()
                if p in triangles:
                    if t_list not in triangles[p]:
                        triangles[p].append(t_list)
                else:
                    triangles[p]=[]
                    triangles[p].append(t_list)
maximised=0
max_p=0
for p , triangle_list in triangles.items():
    if len(triangle_list) > maximised:
        max_p = p
        maximised = len(triangle_list)
print(triangles[max_p])
print(max_p)
```
