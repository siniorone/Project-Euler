```python
from sympy import *
def nontrivial_fraction(numerator,denominator):
    ep=1e-12
    fraction_original=numerator/denominator
    str_nume=str(numerator)
    str_deno=str(denominator)
    for n in str_nume:
       if n!="0" and n in str_deno :
           str_deno=str_deno.replace(n, "",1)
           str_nume=str_nume.replace(n, "",1)
           if int(str_deno)==0 or int(str_nume)==0 : break
           fraction_secondary=int(str_nume)/int(str_deno)
           if fraction_secondary == fraction_original : 
             #print(fraction_original,fraction_secondary)
             return True
           break

    return False
all_fractions=[]
for a in range(10,100):
    for b in range(10,100):
        if a!=b and nontrivial_fraction(a,b) :
           ans=set()
           ans.add(a)
           ans.add(b)
           if ans not in all_fractions:
               all_fractions.append(ans)
print(all_fractions)
product_of_numerator=1
product_of_denominators=1
for f in all_fractions:
    f=list(f)
    if f[0] < f[1]:
        numerator=f[0]
        denominator=f[1]
    else:
        numerator=f[1]
        denominator=f[0]

    
    product_of_denominators *= denominator
    product_of_numerator *= numerator
k = gcd(product_of_denominators, product_of_numerator)
print(product_of_denominators/k)
      
           
```
