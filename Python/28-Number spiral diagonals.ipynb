### Number spiral diagonals
### Problem 28
Starting with the number 1 and moving to the right in a clockwise direction a 5 by 5 spiral is formed as follows:
image.png
It can be verified that the sum of the numbers on the diagonals is 101.

What is the sum of the numbers on the diagonals in a 1001 by 1001 spiral formed in the same way?


---
#### Persian translation:
<div dir='rtl' align='right'>
اگر از عدد یک شروع کنیم و در جهت عقربه های ساعت به راست حرکت کنیم مارپیچ ۵ در ۵ فوق ایجاد می شود 
همانطور که در تصور می بینیم جمع اعداد حاضر در اقطار این مارپیچ برابر با  ۱۰۱ است .
جمع اعداد حاضر در قطرهای یک مارپیچ ۱۰۰۱ در ۱۰۰۱  که با شرایط فوق ایجاد شده باشد را بیابید.
</div>

```python
n = 1001
mysum = 0
d = 2
k = 1
step = 1
while step <= n**2:
    mysum += step
    step += d
    if k == 4: 
       d += 2
       k = 1
    else: 
       k += 1
print(mysum)
```
