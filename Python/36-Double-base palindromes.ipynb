```python
def decimal_to_binary(n):
    return "{0:b}".format(int(n))
decimal={}
decimal[1]=list(range(0,10))
decimal[2]=[11,22,33,44,55,66,77,88,99]
for i in range(3,7):
    if i%2==0:
      decimal[i]=[]
      for x in range(10**(int(i/2)-1),10**(int(i/2))):
          str_x=str(x)
          decimal[i].append(int(str_x+str_x[-1::-1]))
          
    else:
       decimal[i]=[]
       for x in decimal[i-1]:
           str_x=str(x)
           a=str_x[:int(i/2)]
           for y in decimal[1]:
               str_y=str(y)
               decimal[i].append(int(a+str_y+a[-1::-1]))            
all={}
for digit,numbers in decimal.items():
    for n in numbers:
        palindrome=True
        bin=decimal_to_binary(n)
        lbin=len(bin)
        step=int(lbin/2)
        for b in range(step):
            if bin[b]!=bin[-b-1]:
                palindrome=False
                break
        if palindrome: all[n]=bin
print(sum(all.keys()))
        
```
