```python
n=0
while True:
   n+=1 
   digits_n=set(str(n))
   digits_2n=set(str(2*n))
   if digits_2n != digits_n:continue
   digits_3n=set(str(3*n))
   if digits_3n != digits_n:continue
   digits_4n=set(str(4*n))
   if digits_4n != digits_n:continue
   digits_5n=set(str(5*n))
   if digits_5n != digits_n:continue
   digits_6n=set(str(6*n))
   if digits_2n != digits_n:continue
   print(n,"-->",2*n,3*n,4*n,5*n,6*n)
   break
```
