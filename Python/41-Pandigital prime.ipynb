```python
import itertools
def is_prime(n):
    p = 2
    while p < n**.5 :
        if n % p: p +=1 
        else : return False
    return True
condition=True
digits=["1","2","3","4","5","6","7","8","9"]
while condition:
    pandigitals=list(itertools.permutations(digits))
    pandigitals.sort(reverse=True)
    for p in pandigitals:
        p_int=int(''.join(p))
        if is_prime(p_int):
            print(p_int)
            condition=False
            break
    digits.pop(-1)
    
```
