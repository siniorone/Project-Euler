```python
def fact(n):
    if n == 0 : return 1
    if n == 1 : return 1
    return n * fact(n-1)
fact_dict={}
for i in range(101):
    fact_dict[i]=fact(i)
limit=int(1e6)
count=0
for n in range(1,101):
    for r in range(0,n):
        if fact_dict[n]/(fact_dict[n-r]*fact_dict[r]) > limit :
            count += 1
print(count)
```
