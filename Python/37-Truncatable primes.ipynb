```python
def truncate(n):
    result=set()
    str_n=str(n)
    nlen=len(str_n)
    for i in range(len(str_n)):
        result.add(int(str_n[:i+1]))
        result.add(int(str_n[nlen-i-1:]))
    return result
# Find All primes number below 1e6---------------
n = int(1e6)
nonprimes=set()
primes = set()
for i in range(2,n):
    if i in nonprimes:continue
    else:
        primes.add(i)
        for x in range(i+i, n,i):
            if x not in nonprimes: nonprimes.add(x)
# ---------------------------------------------
print(1 in primes)
primes_list=list(primes)
primes_list.sort()
truncatable_primes=[]
count=0
for prime in primes_list:
    if prime < 10 :continue
    truncatable=True
    for t in truncate(prime):
        if t not in primes:
            truncatable=False
            break
    if truncatable :
        truncatable_primes.append(prime)
        count+=1
        if count==11:break
print(truncatable_primes)
print(sum(truncatable_primes))
       
```
