```python
def fact(n):
    if n==0: return 1
    if n==1: return 1
    return n * fact(n-1)
factoriel_dictionary={}
for i in range(0,10):
    factoriel_dictionary[i]=fact(i)
curious_number=set()
for n in range(10,factoriel_dictionary[9]*6+2):
    m=n
    mysum=0
    for i in range(len(str(m))):
        mysum += factoriel_dictionary[m%10]
        m=int(m/10)
    if mysum == n : curious_number.add(n)
print(curious_number)
print(sum(curious_number))
```
