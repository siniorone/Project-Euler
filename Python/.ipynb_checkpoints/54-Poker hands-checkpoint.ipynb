```python
#-----------------------------------------
def is_royal_flush(hand):
    hand_set = set()
    suit = hand[0][-1]
    for h in hand:
        if h[-1] != suit:
            return False
        hand_set.add(h[:-1])
    if hand_set == {'T','J','Q','K','A'}:
        return 100000
    return False
#-----------------------------------------
def high_card_score(hand):
    myhand = [h[:-1] for h in hand]
    card_list = ['2','3','4','5','6','7','8','9','T','J','Q','K','A']
    score = 0
    for i in range(13):
        c = card_list[i]
        if c in myhand:
            score += 2**i
    return score
#-----------------------------------------
def is_straight(hand):
    myhand = set([h[:-1] for h in hand])
    card_list = ['2','3','4','5','6','7','8','9','T','J','Q','K','A']
    for i in range(9):
        ihand = set(card_list[i:i+5])
        if ihand == myhand :
            return 40000
    return False
#-----------------------------------------
def is_flush(hand):
    suit = hand[0][-1]
    for h in hand:
        if h[-1] != suit: return False
    return 50000
#-----------------------------------------
def is_four_of_a_kind(hand,hand_dict):
    score = 0
    hand.sort()
    kind = hand[0][:-1]
    kind_1 = hand[0][:-1]
    remain = []
    count=0
    for h in hand:
        if h[:-1] == kind :
            count += 1
            if count == 4:
                score = 70000
                if kind == kind_1 :
                      remain = hand_dict[hand[0][:-1]] + hand_dict[hand[4][:-1]]
                else :
                      remain = hand_dict[hand[4][:-1]] + hand_dict[hand[0][:-1]]
                score += int(remain)
                return score
        else:
            kind = h[-1]
            count = 1
    return False  
#-----------------------------------------  
def is_full_house(hand,hand_dict):
    hand.sort()
    myhand = set([h[:-1] for h in hand])
    if len(myhand) == 2 :
        hand_3 = hand[2]
        if hand_3 == hand[0] : 
            hand_2 = hand[-1]
        else:
            hand_2 = hand[0]
        remain = hand_dict[hand_3[:-1]] + hand_dict[hand_2[:-1]]
        return 60000 + int(remain)
    return False
#-----------------------------------------
def is_three_of_kind(hand,hand_dict):
    hand.sort()
    kind_3 = hand[2][:-1]
    myhand = []
    for h in hand: myhand.append(h[:-1])
    if myhand.count(kind_3) != 3 : return False
    myhand = []
    for h in hand:
        if h[:-1] != kind_3 :myhand.append(h)
    if int(hand_dict[myhand[0][:-1]]) > int(hand_dict[myhand[1][:-1]]) :
          kind_1 = hand_dict[myhand[0][:-1]]
          kind_2 = hand_dict[myhand[1][:-1]]
    else:
          kind_1 = hand_dict[myhand[1][:-1]]
          kind_2 = hand_dict[myhand[0][:-1]]
    return 30000 + float(hand_dict[kind_3] + kind_1 + "." + kind_2)  
#-----------------------------------------
def is_two_pairs(hand,hand_dict):
    hand.sort()
    myhand = []
    two_pair = []
    one_pair = []
    for h in hand:myhand.append(h[:-1])
    for h in hand_dict : 
        if myhand.count(h) == 2 : two_pair.append(h)
    for h in hand_dict : 
        if myhand.count(h) == 1 : one_pair.append(h)
    if len(two_pair) != 2 : return False
    return 20000 + float(hand_dict[two_pair[1]] + hand_dict[two_pair[0]] + "." + hand_dict[one_pair[0]])
#-----------------------------------------
def is_one_pairs(hand,hand_dict):
    hand.sort()
    myhand = []
    one_pair = []
    no_pair = []
    for h in hand:myhand.append(h[:-1])
    for h in hand_dict : 
        if myhand.count(h) == 2 : one_pair.append(h)
    for h in hand_dict : 
        if myhand.count(h) == 1 : no_pair.append(h)
    if len(one_pair) != 1 : return False
    return 10000 + float(hand_dict[one_pair[0]] + hand_dict[no_pair[2]] + "." + hand_dict[no_pair[1]] + hand_dict[no_pair[0]])  
#-----------------------------------------
def score_my_hand(hand):
    hand_dict={'2':'02','3':'03','4':'04',
               '5':'05','6':'06','7':'07',
               '8':'08','9':'09','T':'10',
               'J':'11','Q':'12','K':'13','A':'14'}
    #-----------------------------------------------
    royal_flush = is_royal_flush(hand)
    if royal_flush : return royal_flush
    #-----------------------------------------------
    flush = is_flush(hand)
    straight = is_straight(hand)
    if flush and straight : return 90000 + high_card_score(hand)
    #-----------------------------------------------
    four_of_a_kind = is_four_of_a_kind(hand,hand_dict)
    if four_of_a_kind :return four_of_a_kind
    #-----------------------------------------------
    full_house = is_full_house(hand,hand_dict)
    if full_house : return full_house
    #-----------------------------------------------
    if flush : return flush + high_card_score(hand)
    #-----------------------------------------------
    if straight: return straight + high_card_score(hand)
    #-----------------------------------------------
    three_of_kind = is_three_of_kind(hand, hand_dict)
    if three_of_kind : return three_of_kind
    #-----------------------------------------------
    two_pairs = is_two_pairs(hand, hand_dict)
    if two_pairs : return two_pairs
    #-----------------------------------------------
    one_pairs = is_one_pairs(hand, hand_dict)
    if one_pairs : return one_pairs
    #-----------------------------------------------
    high_card = high_card_score(hand)
    return  high_card
#----------------------------------------------
f = open("dataset/p054_poker.txt", "r")
all_hands = f.read()
all_hands.replace("\r"," ")
all_hands = all_hands.split()
first_player = 0
second_player = 0
for i in range(0 ,len(all_hands), 10):
    if score_my_hand(all_hands[i:i+5])  >  score_my_hand(all_hands[i+5:i+10]):
        first_player += 1
    else:
        second_player += 1
print(first_player , second_player)
        
```
